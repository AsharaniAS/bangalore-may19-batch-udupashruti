```python
#Question No. 4
laliga_df_goalDifferences = laliga_df.copy()
laliga_df_goalDifferences['GoalsFor'] = laliga_df_goalDifferences['GoalsFor'].astype('int64')
laliga_df_goalDifferences['GoalsAgainst'] = laliga_df_goalDifferences['GoalsAgainst'].astype('int64')
laliga_df_goalDifferences.info()
```

```python
def Goal_diff_count ( df ) :
    diff = abs(df['GoalsFor'] - df['GoalsAgainst'])
    index_max = diff.idxmax()
    print("The team that has maximimum goal difference of ",max(diff),"is ",df.iloc[index_max,1],".")
    index_min = diff.idxmin()
    print("The team that has minimimum goal difference of ",min(diff),"is ",df.iloc[index_min,1],".")
    return diff
```

```python
Goal_diff_count(laliga_df_goalDifferences)
```
