```python
num_texts = data["text"].size
num_texts
```

```python
cleaned_texts = []
```

```python
from bs4 import BeautifulSoup
import re
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
for i in range( 0, num_texts):
    text = BeautifulSoup(data['text'][i]).get_text()
    letters_only = re.sub("[^a-zA-Z]", " ", text)
    words = letters_only.lower().split()
    stops = set(stopwords.words("english"))
    meaningful_words = [w for w in words if not w in stops]
    texts = " ".join( meaningful_words )
    cleaned_texts.append(texts)
```

```python
label_counts = dict()

for labels in df1['labels'].values:
    for label in labels:
        if label in label_counts:
            label_counts[label] += 1
        else:
            label_counts[label] = 1
```

```python
len(label_counts)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer(classes=sorted(label_counts.keys()))
train_labels = mlb.fit_transform(y_train)
```

```python
test_labels = mlb.transform(y_test)
```
