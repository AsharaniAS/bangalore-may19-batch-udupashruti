### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
#Sepal Length (in cm) Sepal Width in (cm) Petal length (in cm) Petal width (in cm) SepalExtra
RangeSepalLength = Iris['Sepal Length (in cm)'].max() - Iris['Sepal Length (in cm)'].min()
print("Range of Sepal Length :", RangeSepalLength)
RangeSepalWidth = Iris['Sepal Width in (cm)'].max() - Iris['Sepal Width in (cm)'].min()
print("Range of Sepal Width :", RangeSepalWidth)
RangePetalLength = Iris['Petal length (in cm)'].max() - Iris['Petal length (in cm)'].min()
print("Range of Petal Length :", RangePetalLength)
RangePetalWidth = Iris['Petal width (in cm)'].max() - Iris['Petal width (in cm)'].min()
print("Range of Petal Width :", RangePetalWidth)
RangeSepalExtra = Iris['SepalExtra'].max() - Iris['SepalExtra'].min()
print("Range of SepalExtra :", RangeSepalExtra)
```
